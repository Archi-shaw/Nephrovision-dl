```python
def get_products(x:int,y:int) -> int:
    return x*y
```

```python
get_products(x=2, y=3)
```

```python
get_products(x=2, y="3")
```

```python
from ensure import ensure_annotations
@ensure_annotations
def get_products(x:int,y:int) -> int:
        return x*y
get_products(x=2, y=3)
get_products(x=2, y="3")
```
